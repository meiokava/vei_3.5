# Индивидуальное задание 3:

<font size="4">Осуществляется распределение оценок в классе и записи количества учащихся, получивших каждую оценку. Нужно визуализировать эти данные с помощью круговой диаграммы. В частности, пусть grades - возможные оценки (например, A, B, C, D, F), а count - количество учащихся, получивших каждую оценк. Задача - создать круговую диаграмму, которая показывает процент учащихся, получивших каждую оценку.

1. Для решения воспользуемся круговой функцией из matplotlib для создания круговой диаграммы. Функция pie принимает данные и метки в качестве входных данных, а также необязательные аргументы для цветов и других свойств клиньев.

2. Используем функцию легенды, чтобы добавить легенду к диаграмме, которая показывает метки и цвета клиньев.
 
3. Определяются входные данные в виде списков оценок и значений подсчета. Затем создается круговая диаграмма процентного соотношения учащихся, получивших каждую оценку, используя функцию pie со значениями count в качестве данных и значениями оценок в качестве меток.

4. Определяются цвета клинов, используя список названий цветов, и форматируются процентные значения, отображаемые в каждом клине, используя аргумент autopct.

5. Добавляется легенда к диаграмме с помощью функции legend, которая отображает метки и цвета клиньев, и отображается график с помощью функции show.

```python
import matplotlib.pyplot as plt

# input data
grades = ['A', 'B', 'C', 'D', 'F']
count = [10, 20, 15, 5, 2]

# create a pie chart of the percentage of students who received each grade
plt.pie(count, labels=grades, colors=['green', 'limegreen', 'yellow', 'orange', 'red'], autopct='%1.1f%%')

# add a legend to the chart
plt.legend()

# display the plot
plt.show()
```
