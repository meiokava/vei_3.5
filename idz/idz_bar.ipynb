# Индивидуальное задание 2:

<font size="4">Изучаются характеристики автомобиля, движущегося с разной скоростью по ровной дороге, было осуществлено измерение времени, необходимое автомобилю, чтобы проехать фиксированное расстояние на каждой скорости. Нужно визуализировать эти данные с помощью столбчатой диаграммы. В частности, пусть speed - это скорость автомобиля в км/ч, а time - время, необходимое автомобилю для прохождения фиксированного расстояния в секундах. Создать столбчатую диаграмму, которая показывает время, необходимое автомобилю для прохождения определенного расстояния в секундах. фиксированное расстояние при каждой скорости, причем по оси x обозначены значения скорости, а по оси y - значения времени.

1. Используется bar из matplotlib для создания столбчатой диаграммы. Функция bar принимает данные x и y в качестве входных данных, а также необязательные аргументы для цвета и ширины полос.

2. Ипользуется функция xticks для установки местоположений отметок по оси x и меток для значений скорости, а функции xlabel и ylabel для добавления меток осей на диаграмму.
    
3. Сначала определяются входные данные в виде списков значений скорости и времени. Затем создается столбчатая диаграмма времени, необходимого автомобилю для прохождения фиксированного расстояния на каждой скорости, используя функцию bar с синими полосами шириной 10 (поскольку значения скорости указаны в единицах км/ч).

4. Затем устанавливается положения отметок по оси x и метки для значений скорости с помощью функции xticks.

5. Добавляются метки осей с помощью функций xlabel и ylabel и отображается график с помощью функции show.

```python
import matplotlib.pyplot as plt

speed = [20, 40, 60, 80, 100]
time = [50, 25, 17, 12, 10]

plt.bar(speed, time, color='blue', width=10)

plt.xticks(speed)

plt.xlabel('Speed (km/h)')
plt.ylabel('Time (s)')

plt.show()
```
