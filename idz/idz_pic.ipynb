<font size="4">Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.

```python
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
from io import BytesIO
import requests

plt.figure(figsize=(25, 10))

response = requests.get('https://i.pinimg.com/564x/27/1d/37/271d37e6b58126200a064865c80be4f0.jpg')
img = Image.open(BytesIO(response.content))
plt.imshow(img)
```
